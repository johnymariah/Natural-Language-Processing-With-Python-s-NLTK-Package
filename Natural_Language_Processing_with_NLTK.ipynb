```python
lemmatizer = nltk.stem.WordNetLemmatizer()
d_lemmatized=[]
for d in df_tokenized:
    lemmatized_token_d2 = [lemmatizer.lemmatize(token) for token in d if token.isalpha()]
    d_lemmatized.append(lemmatized_token_d2)

d_lemmatized
```

### Removing all the stop‐words and the punctuations;

```python
from nltk.corpus import stopwords
d_stopwords_removed = []
for d in d_lemmatized:
    stop_words_removed = [token for token in d if not token in stopwords.words('english') if token.isalpha()]
    d_stopwords_removed.append(stop_words_removed)

print(d_stopwords_removed)
```

###  Based on the output converting each of the reviews in TD‐IDF vectors. The minimal document frequency for each term is 3. Also, including 2‐gram;

```python
#reconstructing document from output of 3
doc_transformed= []
for i in d_stopwords_removed:
    temp = ''
    for j in i:
        temp = temp + j + ' '
    doc_transformed.append(temp)

#vectorizing
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer3 = TfidfVectorizer(ngram_range=(1, 2),min_df=3)
vectorizer3.fit(doc_transformed)
#print(vectorizer3.vocabulary_)
v3 = vectorizer3.transform(doc_transformed)
print(v3.toarray())
```

#### Dimension of TF-IDF result

```python
print(len(vectorizer3.vocabulary_))
```

#### Saving TF-IDF result

```python
import csv
df = pd.DataFrame(v3.toarray(), columns=vectorizer3.vocabulary_)
print(df)
df.to_csv("tf_idf_final.csv", encoding='utf-8', index=False)
```

### Based on the tokenized words in step 1, POS‐tag ing each word and doing a TD‐IDF vectorization, the minimal document frequency for each term is 4 

```python
# 5 POS

#reconstructing the document from 1
df_tokenized_transformed= []
for i in df_tokenized:
    temp = ''
    for j in i:
        temp = temp + j + ' '
    df_tokenized_transformed.append(temp)
    
POS_df_tokenized = []
for doc in df_tokenized_transformed:
    token_doc = nltk.word_tokenize(doc)
    POS_token_doc = nltk.pos_tag(token_doc)
    POS_token_temp = []
    for i in POS_token_doc:
        POS_token_temp.append(i[0] + "_" + i[1])
    POS_df_tokenized.append(" ".join(POS_token_temp))

print(POS_df_tokenized)
```

```python
vectorizer4 = TfidfVectorizer(min_df=4)
vectorizer4.fit(POS_df_tokenized)
#print(vectorizer4.vocabulary_)
POS_v3 = vectorizer4.transform(POS_df_tokenized)
#print(POS_v3.toarray())
```
